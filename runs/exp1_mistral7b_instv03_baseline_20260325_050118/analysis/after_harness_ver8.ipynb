```python
from pathlib import Path
import json
import pandas as pd
import numpy as np
import re

# ✅ run 폴더
run_dir = Path('..').resolve()

print("run_dir exists:", run_dir.exists(), run_dir)
print("traces dir exists:", (run_dir/"traces").exists())
```

```python
from pathlib import Path
import json

pattern = "*part*.json"
json_files = sorted(run_dir.glob(pattern))

print(f"Found {len(json_files)} files")

total = {
    "total_instances": 0,
    "submitted_instances": 0,
    "completed_instances": 0,
    "resolved_instances": 0,
    "unresolved_instances": 0,
    "empty_patch_instances": 0,
    "error_instances": 0,
}

all_resolved_ids = []
all_unresolved_ids = []
all_error_ids = []

for file in json_files:
    with open(file, "r") as f:
        data = json.load(f)

    # total_instances는 max로 처리 (중복 방지)
    total["total_instances"] = max(
        total["total_instances"],
        data.get("total_instances", 0)
    )

    for key in total.keys():
        if key != "total_instances":
            total[key] += data.get(key, 0)

    # ID 수집
    all_resolved_ids.extend(data.get("resolved_ids", []))
    all_unresolved_ids.extend(data.get("unresolved_ids", []))
    all_error_ids.extend(data.get("error_ids", []))  # ✅ 핵심 추가

# 중복 제거
all_resolved_ids = sorted(set(all_resolved_ids))
all_unresolved_ids = sorted(set(all_unresolved_ids))
all_error_ids = sorted(set(all_error_ids))

print("\n=== Aggregated Stats ===")
for k, v in total.items():
    print(f"{k}: {v}")

print("\n=== ID Stats ===")
print(f"resolved_ids: {len(all_resolved_ids)}")
print(f"unresolved_ids: {len(all_unresolved_ids)}")
print(f"error_ids: {len(all_error_ids)}")

print("\nSample error_ids:", all_error_ids[:10])
```
