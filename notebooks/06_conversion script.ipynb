### conversion script
usils.py에 함수 구현 완료

```python
import json
from pathlib import Path

# ✅ 너 실행 폴더로 바꿔
run_dir = Path("../runs/exp1_init_qwen2p5_baseline_20260223_121522")   # 예: runs/exp1_20260223_115230
traces = run_dir / "traces"
out_path = run_dir / "predictions.jsonl"

patch_files = sorted(traces.glob("*_trial*.patch.diff"))

n_written = 0
with out_path.open("w", encoding="utf-8") as f:
    for pf in patch_files:
        # 파일명: {task_id}_trial{trial}.patch.diff
        task_id = pf.name.split("_trial")[0]
        patch = pf.read_text(encoding="utf-8", errors="ignore")

        obj = {
            "instance_id": task_id,
            # 하네스 호환 키들(버전 차이 대비)
            "model_patch": patch,
            "patch": patch,
            "diff": patch,
            "model_name_or_path": "Qwen/Qwen2.5-Coder-7B-Instruct",
        }
        f.write(json.dumps(obj, ensure_ascii=False) + "\n")
        n_written += 1

print("patch files:", len(patch_files))
print("wrote:", out_path, "lines:", n_written)
```
